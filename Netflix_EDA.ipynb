### 2. Comparison of tv shows vs. movies.

```python
data.head()
```

```python
df_MOV = data[data['category'] =='Movie']
```

```python
top_10_Movie_countries = df_MOV.groupby('country')['title'].nunique().sort_values(ascending=False).reset_index().head(10)
top_10_Movie_countries = top_10_Movie_countries.rename(columns={'title': 'Count_of_Movies'})
top_10_Movie_countries
```

```python
df_TV = data[data['category'] =='TV Show']
```

```python
top_10_TV_Show_countries = df_TV.groupby('country')['title'].nunique().sort_values(ascending=False).reset_index().head(10)
top_10_TV_Show_countries = top_10_TV_Show_countries.rename(columns={'title': 'Count_of_TV_Shows'})
top_10_TV_Show_countries
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1,2,1)
plt.bar(top_10_Movie_countries['country'], top_10_Movie_countries['Count_of_Movies'])
plt.title('top_10_Movie_countries')
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('Count of Movies')

plt.subplot(1,2,2)
plt.bar(top_10_TV_Show_countries['country'], top_10_TV_Show_countries['Count_of_TV_Shows'])
plt.title('top_10_TV_Show_countries')
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('Count of TV Shows')

plt.tight_layout()
plt.show()
```

### Insights :
- Here we get to know the country United States was produced more movies and TV shows.
- After US, India was produced more movies and United Kingdom was produced more TV Shows.


### 3. What is the best time to launch Movie or TV show?

a. Find which is the best week to release the Tv-show or the movie. Do the analysis separately for Tv-shows and Movies

```python
data['date_added'].info()
```

```python
data['date_added'] = pd.to_datetime(data['date_added'])
```

```python
data['date_added'].info()
```

```python
data['release_week'] = data['date_added'].dt.isocalendar().week
data.head()
```

```python
df_MOV = data[data['category'] =='Movie']
```

```python
movie_counts = df_MOV.groupby('release_week')['title'].count().reset_index(name = 'Movie_counts')
movie_counts.head(10)
```

```python
df_TV_Show = data[data['category'] =='TV Show']
```

```python
TV_Show_counts = df_TV_Show.groupby('release_week')['title'].count().reset_index(name = 'TV_Show_counts')
TV_Show_counts.head(10)
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(2,1,1)
plt.bar(movie_counts['release_week'], movie_counts['Movie_counts'], color='green')
plt.title('Number of Movies Released Each Week')
plt.xlabel('Release Week')
plt.ylabel('Number of Movies')

plt.subplot(2,1,2)
plt.bar(TV_Show_counts['release_week'], TV_Show_counts['TV_Show_counts'], color='red')
plt.title('Number of TV Shows Released Each Week')
plt.xlabel('Release Week')
plt.ylabel('Number of TV Shows')

plt.tight_layout()
plt.show()
```

```python
Best_Week_to_realse_movie = movie_counts.sort_values(by='Movie_counts', ascending=False)
Best_Week_to_realse_movie.head(3)
```

```python
Best_Week_to_realse_TV_show = TV_Show_counts.sort_values(by='TV_Show_counts', ascending=False)
Best_Week_to_realse_TV_show.head(3)
```

b. Find which is the best month to release the Tv-show or the movie. Do the
analysis separately for Tv-shows and Movies

```python
data['release_month'] = data['date_added'].dt.month
data.head()
```

```python
df_MOV = data[data['category'] =='Movie']
```

```python
movie_counts = df_MOV.groupby('release_month')['title'].count().reset_index(name = 'Movie_counts')
movie_counts
```

```python
df_TV_Show = data[data['category'] =='TV Show']
```

```python
TV_Show_counts = df_TV_Show.groupby('release_month')['title'].count().reset_index(name = 'TV_Show_counts')
TV_Show_counts
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1,2,1)
plt.bar(movie_counts['release_month'], movie_counts['Movie_counts'], width=0.5,color='blue')
plt.title('Number of Movies Released Each Month')
plt.xlabel('Month')
plt.xticks(movie_counts['release_month'], ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.ylabel('Number of Movies')

plt.subplot(1,2,2)
plt.bar(TV_Show_counts['release_month'], TV_Show_counts['TV_Show_counts'], width=0.5, color='orange')
plt.title('Number of TV Shows Released Each Month')
plt.xlabel('Month')
plt.xticks(movie_counts['release_month'], ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.ylabel('Number of TV Shows')

plt.tight_layout()
plt.show()
```

```python
Best_Week_to_realse_movie = movie_counts.sort_values(by='Movie_counts', ascending=False)
Best_Week_to_realse_movie.head(3)
```

```python
Best_Month_to_realse_TV_show = TV_Show_counts.sort_values(by='TV_Show_counts', ascending=False)
Best_Month_to_realse_TV_show.head(3)
```

### Insights :
- The month with the highest number of TV show releases in November and Movies in June.
- June and November also have relatively high TV show and movie releases. This might suggest a pattern of increased TV show and movie releases during the summer months.

### Recommendations :
- Highly-rated movies and TV shows should be released in November and during the summer season to increase customer counts.
